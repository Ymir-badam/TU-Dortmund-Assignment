```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sandwich.csv")

df['antCount'] = pd.to_numeric(df['antCount'], errors='coerce').fillna(0)

mean_bread_data = df.groupby('bread')['antCount'].mean().reset_index()
mean_bread_data.rename(columns={'antCount': 'meanAnts'}, inplace=True)

print("Mean Ant Count by Bread:")
print(mean_bread_data)

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x='bread', y='meanAnts', data=mean_bread_data, palette='viridis')
plt.title('Mean Ant Count by Bread Type', fontsize=16, pad=20) # Updated title
plt.xlabel('Bread Type', fontsize=12)
plt.ylabel('Mean Ant Count', fontsize=12) 
plt.xticks(rotation=45, ha='right') 
for index, row in mean_bread_data.iterrows():
    plt.text(index, row['meanAnts'] + 0.1,
             f"{row['meanAnts']:.2f}", 
             color='black', ha="center", fontsize=10)
plt.tight_layout() 
plt.show() 


mean_topping_data = df.groupby('topping')['antCount'].mean().reset_index()
mean_topping_data.rename(columns={'antCount': 'meanAnts'}, inplace=True)

print("\nMean Ant Count by Topping:")
print(mean_topping_data)


plt.figure(figsize=(10, 6))
sns.barplot(x='topping', y='meanAnts', data=mean_topping_data, palette='magma') # Updated y and data
plt.title('Mean Ant Count by Topping Type', fontsize=16, pad=20) # Updated title
plt.xlabel('Topping Type', fontsize=12)
plt.ylabel('Mean Ant Count', fontsize=12)
plt.xticks(rotation=45, ha='right')

for index, row in mean_topping_data.iterrows():
    plt.text(index, row['meanAnts'] + 0.1,
             f"{row['meanAnts']:.2f}",
             color='black', ha="center", fontsize=10)
plt.tight_layout()
plt.show()


mean_butter_data = df.groupby('butter')['antCount'].mean().reset_index()
mean_butter_data.rename(columns={'antCount': 'meanAnts'}, inplace=True)

print("\nMean Ant Count by Butter Presence:")
print(mean_butter_data)

plt.figure(figsize=(8, 5))
sns.barplot(x='butter', y='meanAnts', data=mean_butter_data, palette='cividis') # Updated y and data
plt.title('Mean Ant Count by Butter Presence', fontsize=16, pad=20) # Updated title
plt.xlabel('Butter (yes/no)', fontsize=12)
plt.ylabel('Mean Ant Count', fontsize=12) 
plt.xticks(rotation=0)
for index, row in mean_butter_data.iterrows():
    plt.text(index, row['meanAnts'] + 0.1, 
             f"{row['meanAnts']:.2f}",
             color='black', ha="center", fontsize=10)
plt.tight_layout()
plt.show()
```
